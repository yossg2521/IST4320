```python
#Q1
import yt_dlp

def get_videoinfo(url):
    #Create a dictionary to stor video info
    video_info = {'name': '', 'description': '', 'uploader': ''}

    #Create an option for yt-dlp to skip video download and only get information
    ydl_opts = {
        'quiet': True,
        'skip_download': True
    }

    #Get info with yt-dlp
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=False)
        video_info['name'] = info.get('title', '')
        video_info['description'] = info.get('description', '')
        video_info['uploader'] = info.get('uploader', '')
    
    return video_info

def output(video_info):
    print(f"Name: {video_info['name']}")
    print(f"Description: {video_info['description']}")
    print(f"Uploader: {video_info['uploader']}")

#Main code
url = input("Please enter a YouTube url: ")
video_info = get_videoinfo(url)
output(video_info)
```

```python
#Q2
import yt_dlp
import re

def get_videoinfo(url):
    #Create a dictionary to stor video info
    video_info = {'name': '', 'description': '', 'uploader': ''}

    #Create an option for yt-dlp to skip video download and only get information
    ydl_opts = {
        'quiet': True,
        'skip_download': True
    }

    #Get info with yt-dlp
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=False)
        video_info['name'] = info.get('title', '')
        video_info['description'] = info.get('description', '')
        video_info['uploader'] = info.get('uploader', '')
    
    return video_info

def output(video_info):
    print(f"Name: {video_info['name']}")
    print(f"Description: {video_info['description']}")
    print(f"Uploader: {video_info['uploader']}")

#Q2 Here's the function for question 2
def get_chapters(text):
    #Use a RegEx pattern to match timestamps and chapter titles
    pattern = r'((?:\d{1,2}:)?\d{1,2}:\d{2})\s+(.+)'

    #Find all matches of the pattern
    matches = re.findall(pattern, text)

    return matches


#Main code
url = input("Please enter a YouTube url: ")
video_info = get_videoinfo(url)
output(video_info)

#Extract the chapters from the description
chapters = get_chapters(video_info['description'])
print("\nChapters found:")
for chapter in chapters:
    print(f"(Timestamp: {chapter[0]}, Title: {chapter[1]})")
```
